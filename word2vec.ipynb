```python
def clean_text(text):
    text = BeautifulSoup(text, 'html.parser').get_text()
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    words = text.lower().split()
    words = [w for w in words if w not in set(stopwords.words('english'))]
    return ' '.join(words)
```

```python

clean_text(df['review'][2])
```

```python
df['clearn_word'] = df.review.apply(clean_text)
```

```python
df.head()
```

```python
vect = CountVectorizer(max_features=5000)
train_data_features = vect.fit_transform(df.clearn_word).toarray()
train_data_features.shape
```

```python
def split_sns(review):
    raw_sentences = nltk.data.load('tokenizers/punkt/english.pickle').tokenize(review.strip())
    print(raw_sentences)
    sentences = [clean_text(s) for s in raw_sentences if s]
    return sentences
```

```python
sentences = sum(df.clearn_word.apply(split_sns), [])
```

```python
sentences
```

```python
sentences_list = []
for line in sentences:
    sentences_list.append(nltk.word_tokenize(line))
```

```python
sentences_list
```

```python
from gensim.models.word2vec import Word2Vec
```

```python
modle = Word2Vec(sentences_list,size=300, window=10,workers=4)
modle.save('D:/mr/models')
```
