```python
import pandas as pd
import jieba
import warnings
import jieba.posseg as pseg
```

```python
data = pd.read_csv('D:/mr/user_tag_query.10W.TRAIN-1w.csv')
data
```

```python
def stopword():
    stop_words = set()
    with open('D:\mr\stopwords.txt', "r", encoding="utf-8") as f:
        for word in f.readlines():
            stop_words.add(word.strip())
    return stop_words
```

```python
words = pseg.cut(''.join(data.QueryList[2].split('\t')))
all_list = list()
allowPOS = ['n','v','j']
for word,flag in words:
    if flag in allowPOS and len(word) > 1:
        all_list.append(word)
```

```python
' '.join([word for word in all_list if word not in stopword()])
```

```python
def clean_text(text):
    words = pseg.cut(''.join(text.split('\t')))
    all_list = list()
    allowPOS = ['n','v','j']
    for word,flag in words:
        if flag in allowPOS and len(word) > 1:
            all_list.append(word)
    return ' '.join([word for word in all_list if word not in stopword()])
```

```python
data['clearn_word'] = data.QueryList.apply(clean_text)
```
